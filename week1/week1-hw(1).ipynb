The networking of data can be represented with three layers that are connected to each other. The first layer being the link layer, which purpose is to connect to the other machine. The second layer is network layer, and its purpose is to create a route to the other end, basically it connects the links together to create a route for the information to move through. The third and last layer is end-to-end layer which purpose is to use the information, for example in a phone call the data would be used to create sound.

**5. Analyze a queue system similar to the tutorial considering $\lambda = 3$ and $\mu = 3$ for 2000 time periods.**

```python
import numpy as np 
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
# 3.)
size = 2000
mean_1 = 3
mean_2 = 4
x = 0
y = 0
arrival = np.random.poisson(mean_1, size)
service = np.random.poisson(mean_2, size)
queue = np.zeros(size+1)

for i in range(size):
    if ((arrival[i] - service[i]) + queue[i]) > 0:
        queue[i+1] = max(0,(arrival[i] - service[i]) + queue[i])
    
    if (queue[i] == 0):
        x = x + 1  

plt.figure(figsize=(14,5))
plt.hist(queue)
plt.xlabel('$n$')
plt.ylabel('Size')
plt.title('Queue')
plt.grid(True)
plt.show() 

print('Worst delay:', np.max(queue))
print("Number of idle periods:", x)
ratio = (x/size)*100
print("Ratio in percents:", ratio, "%")
```
