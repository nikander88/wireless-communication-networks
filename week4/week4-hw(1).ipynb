**2. (Interference models) Explain with your own words the following models and their differences** 

- Collision channel model

The collision channel model expects that if a collision happens, all the data is lost. So it is kind of all or nothing model.

- Capture channel model

The capture channel model expects that if a collision happens, the data can be still received if the signal is strong enough. 

- Vulnerability circle capture model 

The Vulnerability circle capture model expects that the signal can be received if terminal is inside the receiver power area (inside the vulnerability circle) and if there are other terminals active inside the vulnerability circle, then interference can occur. Bv determines the area of the vulnerability circle.

- Power capture model

The power capture model is similar to Vulnerability circle capture model, but it takes into account terminals that are outside the vulnerability circle.

```python
# 1. b.)

a = [2, 3, 4, 5]
d = [2, 3, 4, 5]
P = []
Q = []
W = []
E = []
R = []
for i in range (len(a)):
    Q.append(d[0] ** (-a[i]))
    W.append(d[1] ** (-a[i]))
    E.append(d[2] ** (-a[i]))
    R.append(d[3] ** (-a[i]))


plt.figure(figsize=(14,8))
plt.plot(a, Q, label='d = 2')
plt.plot(a, W, label='d = 3')
plt.plot(a, E, label='d = 4')
plt.plot(a, R, label='d = 5')
plt.legend()
plt.show()
```
